# **Loops**
A "for" loop is used to track a sequence like a list, a dictionary or a string, the keyword for in other programming languages is object-oriented, but let's focus on Python.

To have a direct angle, what you want to execute is a set of declarations, in the form that has been chosen, such as: list.

```python
color = ["red", "verde", "naranja"]
for x in color:
  if x == "verde":
    break
  print(x)
```

The loops are very good in areas where the conditionals are repetitive, they perform repetitive actions until they are fulfilled or not fulfilled, they are useful when that particular action has to be repeated many times.
